# 4차 작업
1. 시군구 -> 주소, 시도, 시군구, 읍면동 으로 변환

```python
import pandas as pd

data = pd.read_csv('./processed/preprocessed_2013_3.csv')
data.head(3)
```

```python
original = data['시군구'].map(lambda x:x.split())
original
```

```python
lst = original.to_list()

city = pd.Series(name='시도',data=list(map(lambda x:x[0],lst)))
gu = pd.Series(name='구',data=list(map(lambda x:x[1],lst)))
district = pd.Series(name='읍면동',data=list(map(lambda x:x[2],lst)))
```

```python
df_new = pd.concat([data,city,gu,district],axis=1)
df_new.to_csv('./processed/preprocessed_2013_4.csv',index=False)
```
