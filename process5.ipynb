```python
import pandas as pd

df = pd.read_csv('./processed/preprocessed_2013_4.csv')
```

```python
df.head(3)
```

```python
date = df[['계약년','계약월','계약일']]
date
```

```python
date_list = date.to_numpy().astype(str).tolist()
new_col = pd.Series(name='계약날짜',data=pd.to_datetime(list(map(lambda x:'-'.join(x), date_list))))

df_new_col = pd.concat([df,new_col],axis=1)
df_new_col.head(3)
```

```python
df_new_col.info()
```

```python
df_new_col.to_csv('./processed/preprocessed_2013_5.csv')
```
