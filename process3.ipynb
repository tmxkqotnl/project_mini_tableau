1. 계약년월 date type으로

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./processed/preprocessed_2013_2.csv')
df.head(3)
```

```python
def insert_point_into_str(x:str):
    lst = list(x)
    lst.insert(4,'.')
    
    return ''.join(lst)
    
year_month = pd.to_datetime(df['계약년월'].astype(str).map(insert_point_into_str),format='%Y-%m')
year = year_month.dt.year
month = year_month.dt.month

print(year,end='\n\n')
print(month,end='\n\n')
```

```python
df_ym = pd.DataFrame({'계약년':year,'계약월':month})

processed = pd.concat([df,df_ym],axis=1)
processed.loc[:,'계약년월'] = year_month.dt.to_period('M')

processed.to_csv('./processed/preprocessed_2013_3.csv',index=False)
```
